```python
# Note! Make sure to load your saved DF before running this function!

def tweetrehydrate2016(source_df, save_df):
    alreadygotitlist = save_df.tweet_id.tolist()
    working_df = source_df[~source_df.tweet_id.isin(alreadygotitlist)] 
    sample = working_df.tweet_id.sample(n = 300, replace = False, random_state= 14)
    id_list = sample.to_list()
    hydro_df = save_df
    tweet_tuple_list=[]
    for i in id_list:
        tweet = test.get_tweet(i)
        tid = i
        tweet_tuple_list.append(({'tweet_id': tid,'tweet_text': str(tweet.data)}))  
    hydro_df = hydro_df.append(tweet_tuple_list, ignore_index=True)
    hydro_df.to_csv('tweets_2016')
            
```
